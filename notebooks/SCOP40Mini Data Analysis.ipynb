```python
import pickle
```

```python
from Bio import PDB
import numpy as np
```

```python
with open("../Minidatabase_all_tasks.pkl","r") as f:
    all_tasks = pickle.load(f)
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
ent_sizes = {}
```

```python
def x(ents,label):
    aasizes = []
    for ent in ents[label]:
        if ent in ent_sizes.keys():
            aasizes.append(ent_sizes[ent])
        else:
            parser = PDB.PDBParser()
            pdb_path = "../SCOP40mini/"+ent+".ent" 
            try:
                structure = parser.get_structure(ent, pdb_path)
            except:
                print(ent)
            model = structure[0]
            aasize= 0
            for chain in model.get_list():
                for residue in chain.get_list():
                    try:
                        coords = residue['CA']
                        aasize+=1
                    except:
                        continue
            ent_sizes[ent]=aasize
            aasizes.append(aasize)  
    return aasizes
        
```

```python
data = []
for task,ents in all_tasks.items():
    print(task)
    pos_test = x(ents,"+test")
    neg_test = x(ents,"-test")
    pos_train = x(ents,"+train")
    neg_train = x(ents,"-train")
    data.append((task,pos_test,neg_test,pos_train,neg_train))
```
